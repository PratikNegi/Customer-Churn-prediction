```python
y = df[["Churn"]]
# x = df.drop("Churn", axis=1, inplace=False)
x = df[["Age", "Gender", "Tenure", "MonthlyCharges", "TotalCharges"]]
```

```python
y
```

```python
x["Gender"] = x["Gender"].apply(lambda x : 1 if x =="Female" else 0)
```

```python
y["Churn"]=y["Churn"].apply(lambda x : 1 if x=="Yes" else 0)
```

```python
y["Churn"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train= scaler.fit_transform(x_train)

```

```python
x_train
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
from sklearn.metrics import accuracy_score
def performance(prediction):
    print("The model evaluation is {}".format(accuracy_score(y_test,prediction )))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel= LogisticRegression()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
logmodel.fit(x_train, y_train)
```

```python
prediction=logmodel.predict(x_test)
```

```python
performance(prediction)
```

```python
score= logmodel.score(x_test, y_test)
score
```

```python
from sklearn.neighbors import KNeighborsClassifier
param = {
    'n_neighbors': [2,4,5,7,9],
    "algorithm": [ 'ball_tree', 'kd_tree','brute']
}
knn = GridSearchCV(KNeighborsClassifier(), param, cv=5)
knn.fit(x_train, y_train)
```

```python
knn.best_params_
```

```python
knn.score(x_test, y_test)
```

```python
from sklearn.svm import SVC
param={
    'C': [1.0, 1.2, 1.8, 2.5],
    'kernel' : ['linear', 'poly', 'rbf','sigmoid'],
    'gamma' : ['scale', 'auto']
}
model = GridSearchCV(SVC(), param, cv=5)
model.fit(x_train, y_train)
```
